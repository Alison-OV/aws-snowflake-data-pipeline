```python
import pandas as pd

# Cargar directamente usando solo el nombre del archivo
df = pd.read_csv('sales_data.csv')

# Mostrar los primeros datos
df.head()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
# primeros registros del set de datos

df.head()
```

```python
# cambiar el customer_id a minusculas y reemplazar los valores vacios por '_'

df['customer_id'].str.lower().str.replace(' ', '_')
```

```python
# set de datos

df
```

```python
#Limpieza de fechas

df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")

df
```

```python
# comprobar valores nulos en la tabla
df.isnull().sum()
```

```python
# manejo de valores nulos

df["customer_id"] = df["customer_id"].fillna("UNKNOWN")
df
```

```python
# normalizacin de pases

df["country"] = (
    df["country"]
    .str.upper()
    .str.replace(".", "", regex=False)
    .replace({
        "USA": "USA",
        "U S A": "USA",
        "UNITED STATES": "USA"
    })
)

df
```

```python
# correccin de precios negativos

df["price"] = (
    df["price"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)

df["price"] = pd.to_numeric(df["price"], errors="coerce").astype("float")
```

```python
df["price"].dtype
```

```python
df = df[df["price"].notna()]
```

```python
df[["price"]].head()
```

```python
df = df[df["price"] > 0]
```

```python
df
```

```python
# ventas por pas

df.groupby("country")["price"].sum().sort_values(ascending=False)
```

```python
# data revenue

df["total_amount"] = df["quantity"] * df["price"]

df
```

```python
# data revenue

df["total_amount"] = df["quantity"] * df["price"]

df.groupby("product")["total_amount"].sum().sort_values(ascending=False)
```

```python
# grafico de barras por producto

import matplotlib.pyplot as plt

# Agrupacin y ordenamiento
sales_by_product = (
    df.groupby("product")["total_amount"]
    .sum()
    .sort_values(ascending=False)
)

# Crear grfico
plt.figure(figsize=(10, 6))
sales_by_product.plot(kind="bar")

# Etiquetas y ttulo
plt.title("Total Sales Amount by Product")
plt.xlabel("Product")
plt.ylabel("Total Amount")

# Rotar etiquetas del eje X
plt.xticks(rotation=45, ha="right")

# Ajuste de layout
plt.tight_layout()

plt.show()
```

```python
# grafico de barras por pais

import matplotlib.pyplot as plt

# Agrupacin y ordenamiento
sales_by_product = (
    df.groupby("country")["total_amount"]
    .sum()
    .sort_values(ascending=False)
)

# Crear grfico
plt.figure(figsize=(10, 6))
sales_by_product.plot(kind="bar")

# Etiquetas y ttulo
plt.title("Total Sales Amount by Product")
plt.xlabel("Product")
plt.ylabel("Total Amount")

# Rotar etiquetas del eje X
plt.xticks(rotation=45, ha="right")

# Ajuste de layout
plt.tight_layout()

plt.show()
```
